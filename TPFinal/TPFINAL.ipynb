### **Diccionario de datos**
* **artist.familiarity**: Es una tasa que describe cuan conocido resulta el artista para el oyente.
* **artist.hottest**: Tasa que describe cuan descargado llego a ser en diciembre 2010.
* **artist.id**: Identificador unico del artista.
* **artist.latitude**: Latitud del lugar de origen del artista.
* **artist.location**: Coordenadas del lugar de origen del artista (En todo el dataset es 0).
* **artist.longitud**: Longitud del lugar de origen del artista.
* **artist.name**: Nombre del artista.
* **artist.similar**: Artista similar (nuevamente en el dataset tiene el valor 0 en todas las filas).
* **artist.terms**: Palabras/Términos mas asociados con el artista.
* **artist.terms_freq**: Frecuencia con que se asocia dicho termino.
* **release.id**: Identificador del album donde se estreno la cancion.
* **release.name**: Nombre del album donde se estreno la cancion (Nuevamente solo tiene 0s).
* **song.artist_mbtags**: Palabra clave realacionada con el artista (No contiene mas que 0s).
* **song.artist_mbtags_count**: Cantidad de palabras claves relacionadas con el artista.
* **song.bars_confidence**: Tasa de "confianza" relacionada a cada estrofa.
* **song.bars_start**: Media del tiempo en el que inica cada estrofa, medido en estrofas
