#### Valores ausentes

```python
d_musica.isnull().sum()
```

Notamos que en la columna *'artist.terms'* nos faltarian 5 valores por lo que los reemplazamos con la moda de dicha columna.

```python
moda_artist_terms = d_musica['artist.terms'].mode()[0]
d_musica.loc[d_musica['artist.terms'].isnull(), 'artist.terms'] = moda_artist_terms
```

Mostramos que ya no tenemos valores ausentes

```python
d_musica.isnull().sum()
```

#### Valores atipicos

Graficamos los boxplots y limpiamos las columnas correspondientes

```python
# Definimos la lista de columnas cuya informacion corresponda a variables cuantitativas
cols = ['artist.familiarity',
        'artist.hotttnesss',
        'artist.terms_freq',
        'song.duration',
        'song.hotttnesss',
        'song.tempo']

posicion = 1
plt.figure(figsize=(10,10))
for c in cols:
    plt.subplot(3,2,posicion)
    plt.title(c)
    plt.boxplot(d_musica[c])
    posicion += 1
plt.show()
```

```python
for c in cols:
    q1 = d_musica[c].quantile(0.25)
    q3 = d_musica[c].quantile(0.75)
    riq = q3 - q1
    franja_inferior = q1 - 1.5 * riq
    franja_superior = q3 + 1.5 * riq
    d_musica.loc[d_musica[c]<franja_inferior, c] =  boxplot_stats(d_musica[c])[0]['whislo']
    d_musica.loc[d_musica[c]>franja_superior, c] =  boxplot_stats(d_musica[c])[0]['whishi']

# Para verificar que no existen valores atipicos vamos a graficar los boxplot
posicion = 1
plt.figure(figsize=(10,10))
for c in cols:
    plt.subplot(3,2,posicion)
    plt.title(c)
    plt.boxplot(d_musica[c])
    posicion += 1
plt.show()
```
