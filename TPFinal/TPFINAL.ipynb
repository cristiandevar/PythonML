#### Valores atipicos

Graficamos los boxplots y limpiamos las columnas correspondientes

```python
# Definimos la lista de columnas cuya informacion corresponda a variables cuantitativas
cols = ['artist.familiarity',
        'artist.hotttnesss',
        'artist.terms_freq',
        'song.duration',
        'song.hotttnesss',
        'song.tempo']

posicion = 1
plt.figure(figsize=(10,10))
for c in cols:
    plt.subplot(3,2,posicion)
    plt.title(c)
    plt.boxplot(d_musica[c])
    posicion += 1
plt.show()
```

```python
for c in cols:
    q1 = d_musica[c].quantile(0.25)
    q3 = d_musica[c].quantile(0.75)
    riq = q3 - q1
    franja_inferior = q1 - 1.5 * riq
    franja_superior = q3 + 1.5 * riq
    d_musica.loc[d_musica[c]<franja_inferior, c] =  boxplot_stats(d_musica[c])[0]['whislo']
    d_musica.loc[d_musica[c]>franja_superior, c] =  boxplot_stats(d_musica[c])[0]['whishi']

# Para verificar que no existen valores atipicos vamos a graficar los boxplot
posicion = 1
plt.figure(figsize=(10,10))
for c in cols:
    plt.subplot(3,2,posicion)
    plt.title(c)
    plt.boxplot(d_musica[c])
    posicion += 1
plt.show()
```

```python
grupo = d_musica.groupby('artist.name')['artist.name']
cantidad = len(grupo.count())

# Aumentamos el tamao de la imagen
plt.figure(figsize=(15,10))

# preparamos los subplots y graficamos primero el pie y luego el bar
plt.subplot(2,2,1)
plt.title(f'Hay {cantidad} diferentes artistas')
grupo.count().sort_values(axis=0, inplace=False, ascending=False)[:5].plot.pie()
plt.ylabel('')

x = grupo.count().sort_values(axis=0, inplace=False, ascending=False)[:5].index.values
y = grupo.count().sort_values(axis=0, inplace=False, ascending=False)[:5].values


plt.subplot(2,2,2)
plt.bar(x,y)
plt.title('5 artistas con mas canciones')
plt.xlabel('Nombres')
plt.ylabel('Cantidad de canciones')


plt.subplot(2,2,3)
x = grupo.count().sort_values(axis=0, inplace=False, ascending=True)[:5].index.values
y = grupo.count().sort_values(axis=0, inplace=False, ascending=True)[:5].values


plt.bar(x,y)
plt.title('5 artistas con menos canciones')
plt.xlabel('Nombres')
plt.ylabel('Cantidad de canciones')
plt.show()
```

**Conclusiones**<br>
Al ver los graficos sacamos las siguientes conclusiones:
