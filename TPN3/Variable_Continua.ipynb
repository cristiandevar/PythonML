```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar una muestra de datos de una variable numérica continua
np.random.seed(42)  # Para reproducibilidad
datos_continuos = np.random.normal(loc=5, scale=2, size=100)  # Datos con distribución normal (media=5, desviación estándar=2)
```

```python
# Crear un DataFrame de pandas para el análisis
df = pd.DataFrame(datos_continuos, columns=['Variable_Numerica_Continua'])

df
```

```python
# Calcular la distribución de frecuencia
frecuencia_absoluta = df['Variable_Numerica_Continua'].value_counts(bins=5, sort=False)
frecuencia_relativa = frecuencia_absoluta / len(df)
# Calcular la frecuencia acumulada
frecuencia_acumulada = frecuencia_absoluta.cumsum()
frecuencia_relativa_acumulada = frecuencia_relativa.cumsum()
# Calcular el x medio (punto medio)
x_medio = frecuencia_absoluta.index.mid
```

```python
print("\nDistribución de frecuencia:")
tabla_frecuencia = pd.DataFrame({'Frecuencia Absoluta': frecuencia_absoluta,
                                 'Frecuencia Relativa': frecuencia_relativa,
                                 'Frecuencia Absoluta Acumulada': frecuencia_acumulada,
                                 'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada,
                                 'x medio': x_medio})
tabla_frecuencia.index.name = 'Rango'
tabla_frecuencia
```

```python
# Histograma
plt.hist(df['Variable_Numerica_Continua'], bins=5, edgecolor='black')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma de Variable Numérica Continua')
plt.show()
```

```python
# Diagrama de frecuencia acumulada
plt.plot(frecuencia_acumulada.index.right, frecuencia_acumulada.values, drawstyle='steps-post')
plt.xlabel('Valor')
plt.ylabel('Frecuencia Acumulada')
plt.title('Diagrama de Frecuencia Acumulada de Variable Numérica Continua')
plt.show()
```

```python
# Diagrama de frecuencia relativa acumulada
plt.plot(frecuencia_relativa_acumulada.index.right, frecuencia_relativa_acumulada.values, drawstyle='steps-post')
plt.xlabel('Valor')
plt.ylabel('Frecuencia Relativa Acumulada')
plt.title('Diagrama de Frecuencia Relativa Acumulada de Variable Numérica Continua')
plt.show()
```

```python
# Diagrama de frecuencia relativa acumulada
plt.plot(frecuencia_relativa_acumulada.index.right, frecuencia_relativa_acumulada.values, drawstyle='steps-post')
plt.xlabel('Valor')
plt.ylabel('Frecuencia Relativa Acumulada')
plt.title('Diagrama de Frecuencia Relativa Acumulada de Variable Numérica Continua')
plt.show()
```

```python
# Obtener algunos estadísticos descriptivos
media = df['Variable_Numerica_Continua'].mean()
mediana = df['Variable_Numerica_Continua'].median()
minimo = df['Variable_Numerica_Continua'].min()
maximo = df['Variable_Numerica_Continua'].max()
# medidas de dispersión
desviacion_estandar = df['Variable_Numerica_Continua'].std()
rango = df['Variable_Numerica_Continua'].max() - df['Variable_Numerica_Continua'].min()
```

```python
# Imprimir los estadísticos descriptivos, medidas de dispersión y distribución de frecuencia
print("Media:", media)
print("Mediana:", mediana)
print("Mínimo:", minimo)
print("Máximo:", maximo)
print("Desviación Estándar:", desviacion_estandar)
print("Rango:", rango)
```

```python
# Diagrama de caja (boxplot)
plt.boxplot(df['Variable_Numerica_Continua'], vert=False)
plt.xlabel('Valor')
plt.title('Diagrama de Caja de Variable Numérica Continua')
plt.show()
```
